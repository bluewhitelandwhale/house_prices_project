## Training a linear model, without the missing values

```python
# Select only numerical columns
df_numeric = df.select_dtypes(include=['int64', 'float64'])

# Drop columns with any missing values
df_clean = df_numeric.dropna(axis=0)

print(f"Original shape: {df.shape}")
print(f"Numeric shape: {df_numeric.shape}")
print(f"Clean shape: {df_clean.shape}")
```

## Spliting the data

```python
from sklearn.model_selection import train_test_split, cross_val_score
```

## % missing values

```python
miss_perc = (df.isnull().mean()*100)
miss_perc = miss_perc[miss_perc > 0]
miss_perc.sort_values(ascending=False)
```

```python
none_type = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'MasVnrType', 'FireplaceQu', 'GarageCond', 'GarageType', 'GarageFinish', 'GarageQual', 'BsmtFinType2', 'BsmtExposure',
             'BsmtQual', 'BsmtCond', 'BsmtFinType1', 'GarageYrBlt', 'MasVnrArea']
```

```python
# for LotFrontage since 17% of the data is alot, we dont drop and also none doesn't seem right...
median_type = ['LotFrontage']
```

```python
# A really small portion of the data is missing, just replace with mode (the most frequent value for this feature)
mode_type = ['Electrical']
```

```python
# filling categorical features with 'None' and numerical with 0
for cat_feature in none_type:
    if df[cat_feature].dtype in ['float64', 'int64']:
        df[cat_feature].fillna(0, inplace=True)
    else:
        df[cat_feature].fillna('None', inplace=True)
```

```python
# filling with the median
for median_feature in median_type:
    df[median_feature].fillna(df[median_feature].median(), inplace=True)
```

```python
# filling with the mode
for mode_feature in mode_type:
    df[mode_feature].fillna(df[mode_feature].mode()[0], inplace=True)
```

# Coefficient comparision

```python
# FIX: your feature names are NOT the index — they are a column
# This version forces feature-name alignment and produces sane output

import pandas as pd

# assume first column = feature names, second column = coefficients
c0 = (
    coef_df0
    .iloc[:, :2]
    .set_index(coef_df0.columns[0])
    .iloc[:, 0]
    .pipe(pd.to_numeric, errors="coerce")
    .rename("model_0")
)

c1 = (
    coef_df1
    .iloc[:, :2]
    .set_index(coef_df1.columns[0])
    .iloc[:, 0]
    .pipe(pd.to_numeric, errors="coerce")
    .rename("model_1")
)

coef_compare = (
    pd.concat([c0, c1], axis=1, join="outer")
    .assign(
        abs_diff=lambda df: (df["model_1"] - df["model_0"]).abs(),
        sign_flip=lambda df: (df["model_0"] * df["model_1"]) < 0,
        present_in_both=lambda df: df["model_0"].notna() & df["model_1"].notna()
    )
    .sort_values("model_0", ascending=False)
)

coef_compare
```
