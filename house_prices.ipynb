# Pipelines, Imputers, preprocessing and model

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import Ridge
```

```python
X = df.drop(['SalePrice', 'Id'], axis=1)
y = df['SalePrice']

num_features = X.select_dtypes(include=['int64', 'float64']).columns

# for some features missing value has meaning like Garage year build (house has no garage), can't be replaced with median or mode.
# same with masanory area. so its best to fill with zero.
num_zero_features = ['GarageYrBlt', 'MasVnrArea']

# avoiding overlaping
num_features = num_features.drop(['GarageYrBlt', 'MasVnrArea'])


# cat_features = X.select_dtypes(include=['object']).columns

# categorical features with null value mean doesn't exit, i.e no fence, fireplace or garage
# best to replace with none
cat_none_features = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'MasVnrType', 'FireplaceQu', 'GarageCond', 'GarageType', 'GarageFinish', 'GarageQual', 'BsmtFinType2', 'BsmtExposure',
             'BsmtQual', 'BsmtCond', 'BsmtFinType1']

# few missing values, best to be replaced with mode
cat_mode_features = ['Electrical']
```

```python
num_median_transformer = Pipeline([(
    "imputer", SimpleImputer(strategy='median'))
])
num_zero_transformer = Pipeline([(
    "imputer", SimpleImputer(strategy='constant', fill_value=0))
])

cat_none_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="constant", fill_value="None")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])
cat_mode_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])


preprocessor = ColumnTransformer(
    transformers=[
        ("num_median", num_median_transformer, num_features),
        ("num_zero", num_zero_transformer, num_zero_features),
        ("cat_none", cat_none_transformer, cat_none_features),
        ("cat_mode", cat_mode_transformer, cat_mode_features)
    ]
)

pipeline = Pipeline([
    ('preprocess', preprocessor),
    ('model', Ridge(alpha=10))
])
```

# Finding the best alpha for ridge
manually
alpha=0.01   mean RMSE=36144 | std=8429  
alpha=0.1    mean RMSE=35334 | std=7744  
alpha=1.0    mean RMSE=34664 | std=7529  
alpha=10     mean RMSE=34151 | std=7583  
alpha=100    mean RMSE=34223 | std=7760  

**alpha = 10**
- Mean RMSE is clearly lower
- Std barely increases (still very stable)
- No sign of underfitting yet

```python
alphas = [0.01, 0.1, 1.0, 10, 100]

for a in alphas:
    pipeline = Pipeline([
        ('preprocess', preprocessor),
        ('model', Ridge(alpha=a))
    ])
    
    cv_scores = cross_val_score(
        pipeline,
        X,
        y,
        scoring="neg_root_mean_squared_error",
        cv=5
    )
    
    rmse = -cv_scores
    print(f"alpha={a:<6} mean RMSE={rmse.mean():.0f} | std={rmse.std():.0f}")
```

**alpha = 10**
- Mean RMSE is clearly lower
- Std barely increases (still very stable)
- No sign of underfitting yet

# Extracting features and coefficients from Ridge model

```python
pipeline.fit(X, y)
```

```python
preprocessor = pipeline.named_steps['preprocess']
```

```python
num_features_out = list(num_features) +  list(num_zero_features)
```

```python
cat_none_ohe = preprocessor.named_transformers_["cat_none"] \
    .named_steps["onehot"] \
    .get_feature_names_out(cat_none_features)

cat_mode_ohe = preprocessor.named_transformers_["cat_mode"] \
    .named_steps["onehot"] \
    .get_feature_names_out(cat_mode_features)
```

```python
feature_names = (
    num_features_out +
    list(cat_none_ohe) +
    list(cat_mode_ohe)
)
```

```python
coefs = pipeline.named_steps["model"].coef_
```

```python
# Sanity check
len(feature_names), len(coefs)
```

```python
coef_df = (
    pd.DataFrame({
        "feature": feature_names,
        "coefficient": coefs
    })
    .assign(abs_coef=lambda df: df["coefficient"].abs())
    .sort_values("abs_coef", ascending=False)
)
```
